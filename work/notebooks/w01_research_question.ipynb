## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Search question:** Which content pages should a review team look at first,
out of thousands, when deciding where to spend limited refresh effort?

**Unit of analysis:** a single content page (`content_id`), evaluated over its
recent 90-day window.

**Output:** a ranked queue of pages, each with a score and a reason code
(e.g. "stale visible page," "declining with demand") explaining why it
surfaced.

**The decision this improves:** which pages a content editor or SEO reviewer
opens first this week, out of a large inventory they cannot fully review by hand.

**The action someone takes:** based on the ranked queue and reason code, a
reviewer decides to refresh, expand, protect, or simply monitor a specific page.

**Cost of a wrong call:**
- False positive (flagged as worth reviewing, but not actually declining):
  wastes a reviewer's limited time on a page that didn't need attention.
- False negative (a genuinely declining, high-value page not surfaced):
  the page keeps losing visibility/traffic silently, with no one aware to act.

Given limited review capacity, both errors have a real cost — which is why
Precision@K (not just accuracy) is the right metric: it directly measures
how many of the reviewer's limited slots would actually be worth their time.

**Why data/ML helps here, and why this isn't "just train a model":** a fixed
hand-written rule (the baseline) can only combine a small number of thresholds
a human guessed at. A learned model can weigh many observable signals
(impressions, position, freshness, word count, engagement) jointly and
non-linearly — but only if validated honestly (client-holdout, not in-sample),
and only if it's producing a *ranked, explainable* review queue rather than an
opaque prediction. The real work is defining a clean future-looking label,
auditing for leakage, and turning a probability into an actionable, reviewable
list — not just fitting a classifier.
